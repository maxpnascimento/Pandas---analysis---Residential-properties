## Residential Properties

```python
import pandas as pd
```

```python
data = pd.read_csv("dataset/aluguel.csv", sep = ";")

data
```

```python
list(data['Tipo'].drop_duplicates())
```

```python
residential = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila',
]
residential
```

```python
selection = data['Tipo'].isin(residential)   #O isin(lista) marca com True os elementos que o tipo está na lista de tipos passada

selection
```

```python
residential_data = data[selection]        #Seleciona apenas o elementos marcados com True

residential_data
```

```python
residential_data.index = range(residential_data.shape[0])

residential_data
```

## Exporting dataset

```python
residential_data.to_csv('dataset/aluguel_residencial.csv', sep = ';', index = False)
```
