## Creating Groupings and Statistics

```python
import pandas as pd
```

```python
data =  pd.read_csv("dataset/aluguel_residencial.csv", sep = ';')

data.head(10)
```

```python
data['Valor'].mean()
```

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']

data = data[data['Bairro'].isin(bairros)]

data
```

```python
grupo_bairro = data.groupby('Bairro')    #Create a groups with each neighborhood

grupo_bairro.groups
```

```python
#Visualizando a média de uma variável em cada grupo
for bairro, dados in grupo_bairro:
    print("{} -> {}".format(bairro, dados.Valor.mean()))
```

```python
#Visualizando a média de uma variável em cada grupo de maneira mais rápida
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

## Estatísticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)         #Retorna algumas estatísticas descritivas do df
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum'])
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = "red")
fig.set_ylabel("Valor")
fig.set_xlabel("Bairro")
fig.set_title("Valor médio do aluguel por bairro", {'fontsize': 22})
```

```python
fig = grupo_bairro['Valor'].std().plot.bar(color = "red")
fig.set_ylabel("Valor")
fig.set_xlabel("Bairro")
fig.set_title("Valor Desvio Padrão no valor do aluguel por bairro", {'fontsize': 22})
```
