## Creating new variables

```python
import pandas as pd
```

```python
df = pd.read_csv("dataset/aluguel_residencial.csv", sep =  ';')

df
```

```python
#Valor Bruto (Valor do aluguel + Condomínio + IPTU)

df['Valor Bruto'] = df['Valor'] + df['Condominio'] + df['IPTU']

df.head(10)
```

```python
#Valor por m2

df['Valor m2'] = df['Valor'] / df['Area']

df['Valor m2'] = df['Valor m2'].round(2)    #Arredonda para duas cassa decimais

df.head(10)
```

```python
df['Valor Bruto m2'] = df['Valor Bruto']/df['Area']
df
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']

df['Tipo Agregado'] = df['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')  #.apply( ) permite rodar uma função em cada linha do df

df.head(100)
```

## Excluding variables

```python
df_aux = pd.DataFrame(df[['Tipo Agregado', 'Valor m2', 'Valor Bruto']])

df_aux
```

```python
del df_aux['Valor Bruto']

df_aux
```

```python
df_aux.pop('Valor Bruto')

df_aux
```

```python
df.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
df
```
